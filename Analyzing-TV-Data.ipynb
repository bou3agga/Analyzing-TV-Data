```python
# Import pandas
import pandas as pd

# Load the CSV data into DataFrames
super_bowls = pd.read_csv('datasets/super_bowls.csv')
tv = pd.read_csv('datasets/tv.csv')
halftime_musicians = pd.read_csv('datasets/halftime_musicians.csv')

# Display the first five rows of each DataFrame
display(super_bowls.head())
display(tv.head())
display(halftime_musicians.head())
```

```python
#inspecting the tv data 
tv.info()
print("\n")
#inspecting the halftime_musicians
halftime_musicians.info()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.style.use('seaborn')

#ploting the histogram of combined points

super_bowls['combined_pts']=super_bowls['winning_pts']+super_bowls['losing_pts']
plt.hist(super_bowls['combined_pts'])
plt.xlabel('Combined Points')
plt.ylabel('Number of Super Bowls')
# Displaying the Super Bowls with the highest and lowest combined scores
display(super_bowls[super_bowls['combined_pts']>70])
display(super_bowls[super_bowls['combined_pts']<25])
```

```python
#ploting the histogram of point diffrence 
plt.hist(super_bowls.difference_pts)
plt.xlabel("Point Diffrence ")
plt.ylabel("Number of Super Bowls")
plt.show()
#Super Bowls where the point difference was equal to 1
display(super_bowls[super_bowls['difference_pts']==1])
#Super Bowl(s) where the point difference was greater than or equal to 35
display(super_bowls[super_bowls['difference_pts']>=35])
```

```python
# Join game and TV data
games_tv=pd.merge(tv[tv['super_bowl']>1],super_bowls,on='super_bowl')
# Importing seaborn
import seaborn as sns 
# Creating a scatter plot with a linear regression model fit
sns.regplot(x='difference_pts',y='share_household',data=games_tv)
plt.xlabel('difference points')
plt.ylabel('share household')
```
